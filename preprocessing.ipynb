```python
#decide wheater we need to do Outlier treatments, we can find it by boxplot done above and found outloerded
```

```python
#this is the function which we will use below 

def wisker(col):
    q1,q3= np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
df.columns
```

```python
wisker(df['GDP'])
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()

# Verify the cleaned column names
print("Cleaned columns:", df.columns)
```

```python
for i in ['GDP', 'Total expenditure', 'thinness 1-19 years']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```
